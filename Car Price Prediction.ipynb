```python
for col in df.columns:
    print('Unique values of ' + col)
    print(df[col].unique())
    print("=========================================================================")
```

```python
def get_brand_name(car_name):
    car_name = car_name.split(' ')[0]
    return car_name.strip()
```

```python
def clean_data(value):
    value = value.split(' ')[0]
    value = value.strip()
    if value == '':
        value = 0
    return float(value)
```

```python
df['name'] = df['name'].apply(get_brand_name)
```

```python
df['name'].unique()
```

```python
for col in df.columns:
    print('Unique values of ' + col)
    print(df[col].unique())
    print("=========================================================================")
```

```python
mapping = {
    'Maruti': 1, 'Hyundai': 2, 'Datsun': 3, 'Honda': 4, 'Tata': 5,
    'Chevrolet': 6, 'Toyota': 7, 'Jaguar': 8, 'Mercedes-Benz': 9, 'Audi': 10,
    'Skoda': 11, 'Jeep': 12, 'BMW': 13, 'Mahindra': 14, 'Ford': 15,
    'Nissan': 16, 'Renault': 17, 'Fiat': 18, 'Volkswagen': 19, 'Volvo': 20,
    'Mitsubishi': 21, 'Land': 22, 'Daewoo': 23, 'MG': 24, 'Force': 25,
    'Isuzu': 26, 'OpelCorsa': 27, 'Ambassador': 28, 'Kia': 29
}

df['name'] = df['name'].replace(mapping)
```

```python
df['transmission'].unique()
```

```python
df['transmission'] = df['transmission'].replace(['Manual', 'Automatic'], [1, 2])
```

```python
df['seller_type'].unique()
```

```python
df['seller_type'] = df['seller_type'].replace(['Individual', 'Dealer', 'Trustmark Dealer'], [1, 2, 3])
```

```python
df.info()
```

```python
df['fuel'].unique()
```

```python
df['fuel'] = df['fuel'].replace(['Petrol', 'Diesel', 'CNG', 'LPG', 'Electric'], [1,2,3,4,5])
```

```python
df.info()
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
df['owner'].unique()
```

```python
df['owner'] = df['owner'].replace(['First Owner', 'Second Owner', 'Third Owner', 'Fourth & Above Owner', 'Test Drive Car'], [1,2,3,4,5])
```

```python
df.drop(columns=['index'], inplace=True)
```

```python
for col in df.columns:
    print('------------------------------------------------------------------------')
    print(col)
    print(df[col].unique())
```

```python
df.isnull().sum()
```

```python
input_data = df.drop(columns=['selling_price'])
output_data =df['selling_price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.2)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
```
